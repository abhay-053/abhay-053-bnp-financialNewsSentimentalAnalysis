```python
# import the necessary libraries
import nltk
import pandas as pd
import re
from nltk.corpus import stopwords
from nltk.stem.porter import PorterStemmer
```

```python
nltk.download('stopwords')
```

```python
ps = PorterStemmer()

def stemming(content):
    content = re.sub('[^a-zA-Z]', ' ', content)
    content = re.sub(r'\s+', ' ', content).strip()
    content = content.lower()
    words = content.split()
    stemmed_words = [ps.stem(word) for word in words if word not in stopwords.words('english')]
    processed_content = ' '.join(stemmed_words)
    return processed_content
```

```python
df['Cleaned Text']= df['Article_Content'].apply(stemming)
```

```python
import pandas as pd
from transformers import pipeline
import numpy as np

# Initialize the FinBERT sentiment classifier
model_name = "ProsusAI/finbert"
classifier = pipeline("sentiment-analysis", model=model_name, device=0)

# Read your dataset
df = pd.read_csv('Preprocessed_news.csv')

def sliding_window(text, window_size=512, stride=256):
    """
    Split the text into overlapping chunks using a sliding window.
    
    Args:
        text (str): The text to split.
        window_size (int): The size of the window (maximum token length).
        stride (int): The step between windows.
    
    Returns:
        list: A list of text chunks.
    """
    chunks = []
    for i in range(0, len(text), stride):
        chunks.append(text[i:i + window_size])
        if len(chunks[-1]) < window_size:
            break
    return chunks

def weighted_sentiment_recommendation_with_window(text, chunk_size=512, stride=256):
    """
    Analyze sentiment for long text by dividing it into chunks using a sliding window.
    
    Args:
        text (str): The input text to analyze.
        chunk_size (int): The maximum size of each chunk (in characters).
        stride (int): The stride between each window.
    
    Returns:
        overall_label (str): Sentiment label based on weighted score.
        recommendation (str): Recommendation based on weighted sentiment score.
    """
    # Apply sliding window to break the text into chunks
    chunks = sliding_window(text, chunk_size, stride)
    
    # Initialize weighted score
    weighted_score = 0
    
    # Process each chunk
    for chunk in chunks:
        result = classifier(chunk)[0]
        sentiment_label = result['label']
        sentiment_score = result['score']
        
        # Add weighted score based on sentiment
        if sentiment_label == "positive":
            weighted_score += sentiment_score
        elif sentiment_label == "negative":
            weighted_score -= sentiment_score
    
    # Determine overall sentiment label
    overall_label = "positive" if weighted_score > 0 else "negative"
    
    # Map weighted score to recommendation
    if weighted_score >= 3:
        recommendation = 'Strong Buy'
    elif 1.5 <= weighted_score < 3:
        recommendation = 'Buy'
    elif -1.5 < weighted_score < 1.5:
        recommendation = 'Hold'
    elif -3 < weighted_score <= -1.5:
        recommendation = 'Sell'
    else:
        recommendation = 'Strong Sell'
    
    return overall_label, recommendation

# Apply the function to each row of the 'Cleaned Text' column
df[['Sentiment_Label', 'Recommendation']] = df['Cleaned Text'].apply(
    lambda text: pd.Series(weighted_sentiment_recommendation_with_window(text))
)

# Save or inspect the updated dataset
df.to_csv('Updated_news_with_recommendations.csv', index=False)

# Optionally, inspect the results
print(df[['Sentiment_Label', 'Recommendation']].head())
```
